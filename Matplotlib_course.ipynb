```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plot

```python
var1=np.arange(0,10,2)
var2=var1**2
```

```python
plt.plot(var1)
plt.show()
```

```python
plt.plot(var2)
plt.show()
```

```python
plt.plot(var1,var2)
plt.show()
```

```python
var3=var1**3
```

```python
plt.plot(var1,var2)
plt.plot(var1,var3)
plt.show()
```

```python
plt.plot(var1,var2,label='Var ke-2', marker='x')
plt.plot(var1,var3,label='Var ke-3', marker='o')
plt.legend(loc='best')
plt.xlabel("Jarak")
plt.ylabel("Waktu")
plt.title("Contoh Gambar-1")
plt.grid()
plt.show()
```

## Bar

```python
var1=['a','b','c']
var2=[1,2,3]
plt.bar(var1,var2)
plt.ylabel("Nilai")
plt.grid()
plt.show()
```

```python
label=['var1','var2','var3']
no_label=np.arange(len(label))
varA=[1,2,3]
varB=[3,4,5]
lbr=0.3
plt.bar(no_label,varA, color='gold', width=lbr, label='varA')
plt.bar(no_label+lbr,varB, color='green', width=lbr, label='varB')
plt.grid()
plt.legend(loc='best')
plt.show()
```

## Pie

```python
var_label=['a','b','c']
var_val=[1,2,3]
plt.pie(var_val, labels=var_label)
plt.show()
```

```python
var_lab=['a','b','c']
var_val=[3,2,1]
plt.pie(var_val, labels=var_lab, autopct="%1.2f%%", shadow=True, explode=[0,0.1,0])
plt.show()
```
